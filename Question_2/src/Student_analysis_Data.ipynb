```python
import pandas as pd
```

```python
student_raw_data=pd.read_csv('/content/StudentsPerformance.csv')
```

```python
student_raw_data
```

```python
student_raw_data['gender'].replace(['female', 'male'],
                        [0, 1], inplace=True)
student_raw_data['test preparation course'].replace(['none', 'completed'],
                        [0, 1], inplace=True)

student_raw_data['Total score'] = student_raw_data['math score'] + student_raw_data['writing score'] + student_raw_data['reading score']
```

```python
student_raw_data
```

```python
student_clean_data=student_raw_data.dropna()
student_clean_data
student_clean_data.to_csv('/content/StudentsPerformance_clean_data.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
data=pd.read_csv('/content/StudentsPerformance_clean_data.csv')
```

```python
data
```

```python
#1
figure, axis = plt.subplots(2, 2)


sns.boxplot(x=data['gender'],y=data['math score'],ax=axis[0,0])
sns.boxplot(x=data['gender'],y=data['reading score'],ax=axis[0,1])
sns.boxplot(x=data['gender'],y=data['writing score'],ax=axis[1,0])
sns.boxplot(x=data['gender'],y=data['Total score'],ax=axis[1,1])
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.suptitle('Scores distribution by gender')
plt.savefig('/content/scoresdistribution.png')
```

```python
#2

plt.bar(data['gender'].value_counts().index,
       data['gender'].value_counts().values,
       color = ['pink', 'red'])
plt.title('Gender Distribution')
plt.xticks(data['gender'].value_counts().index,['Female','Male'])
plt.savefig('/content/genderdistribution.png')
```

```python
#3
figure, axis = plt.subplots(2,2)

sns.histplot(data['math score'],ax=axis[0,0],label='math score')
sns.histplot(data['reading score'],ax=axis[0,1],label='reading score')
sns.histplot(data['writing score'],ax=axis[1,0],label='writing score')
sns.histplot(data['Total score'],ax=axis[1,1],label='Total score')
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.suptitle('Frequency distribution of scores')
plt.savefig('/content/scoresfrequency.png')
```

```python
#4

plt.scatter(data['math score'], data['reading score'], c=data['gender'])
plt.legend(['female'])
plt.xlabel('math score')
plt.ylabel('reading score')
plt.title('math score distribution vs reading score distribution')
plt.savefig('/content/mathvsreadingdistribution.png')
```

```python
#5

graph = sns.FacetGrid(data, col ='gender',hue_kws={'color': ['b', 'g']}, hue ='test preparation course')
graph.map(plt.hist, 'race/ethnicity', bins = 15, color ='orange').add_legend()
plt.savefig('/content/coursecompletiondistribution.png')
```
